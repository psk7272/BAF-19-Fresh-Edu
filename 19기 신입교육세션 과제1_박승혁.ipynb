# 19기 신입교육세션 과제1_박승혁

## 관심 있는 주제
저축은행 및 은행 부실 가능성 분석

## 주제를 선택한 이유
금융회사의 안정성과 부실 가능성을 파악할 수 있는 데이터를 선택했습니다. 저축은행의 주요 경영지표와 실제 파산은행 사례를 함께 살펴보면 금융회사의 건전성을 이해하는 데 도움이 될 것이라고 생각했습니다. 특히 BIS자기자본비율, 연체율, 고정이하여신비율 등은 금융회사의 부실 위험을 판단하는 데 중요한 지표이기 때문에 관심 주제로 선정했습니다.

---

## 데이터 1 : 파산 은행 데이터셋 (Failed Banks Dataset)

- 출처 : Kaggle  
  https://www.kaggle.com/datasets/sahirmaharajj/failed-banks-dataset?resource=download

- 분야 : Finance

- 데이터 특징 :  
  이 데이터는 미국 FDIC의 파산은행 목록 데이터입니다. 은행명, 도시, 주, 폐쇄일, 인수기관 등의 정보가 포함되어 있어 실제로 어떤 은행이 언제, 어느 지역에서 파산했는지 확인할 수 있습니다.  
  이를 통해 은행 파산이 특정 시기나 지역에 집중되는지, 파산한 은행들이 어떤 기관에 인수되었는지 등을 분석할 수 있습니다. 실제 파산 사례를 살펴볼 수 있다는 점에서 은행 부실 가능성 분석의 참고 자료로 활용할 수 있습니다.

---

## 데이터 2 : 예금보험공사_저축은행업권 주요 경영지표 현황

- 출처 : 공공데이터포털  
  https://www.data.go.kr/data/15120292/fileData.do

- 분야 : 일반공공행정 - 재정·금융

- 데이터 특징 :  
  이 데이터는 국내 저축은행업권의 주요 경영지표 현황을 제공하는 데이터입니다. 저축은행명, 지역, 총자산, 대출금, 당기순이익, BIS자기자본비율, 연체율, 고정이하여신비율, 점포수 등의 정보가 포함되어 있습니다.  
  특히 BIS자기자본비율은 금융회사의 자본 건전성을 보여주는 지표이고, 연체율과 고정이하여신비율은 대출 부실 가능성을 파악하는 데 활용될 수 있습니다. 따라서 저축은행별 재무 상태와 부실 위험을 비교·분석하는 데 적합한 데이터라고 생각합니다.
